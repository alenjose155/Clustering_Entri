```python

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_iris
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from scipy.cluster.hierarchy import dendrogram, linkage, fcluster


iris = load_iris()
X = pd.DataFrame(iris.data, columns=iris.feature_names)


scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)


kmeans = KMeans(n_clusters=3, random_state=42)
kmeans_labels = kmeans.fit_predict(X_scaled)


plt.figure(figsize=(8, 6))
sns.scatterplot(x=X_scaled[:, 0], y=X_scaled[:, 1], hue=kmeans_labels, palette='viridis', s=50)
plt.title("KMeans Clustering on Iris Dataset")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.legend(title="Clusters", loc='best')
plt.show()


linkage_matrix = linkage(X_scaled, method='ward')
hier_labels = fcluster(linkage_matrix, t=3, criterion='maxclust')


plt.figure(figsize=(20, 20))
dendrogram(linkage_matrix)
plt.title("Dendrogram for Iris Dataset (Hierarchical Clustering)")
plt.xlabel("Sample Index")
plt.ylabel("Distance")
plt.show()

plt.figure(figsize=(8, 6))
sns.scatterplot(x=X_scaled[:, 0], y=X_scaled[:, 1], hue=hier_labels, palette='cool', s=50)
plt.title("Hierarchical Clustering on Iris Dataset")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.legend(title="Clusters", loc='best')
plt.show()
```
